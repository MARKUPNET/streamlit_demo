```python
df = pd.read_excel('data.xlsx', sheet_name='Sheet1')
df.set_index('datetime', inplace=True)
```

```python
df['欠損'] = pd.notna(df['value'])
df['value2'] = df['value'].fillna(0)
df['value2'] = df['value2'].astype('int')
df['差分直前'] = df['value2'].diff(1)
df['差分直後'] = df['value2'].diff(-1)

df.to_excel('result.xlsx', sheet_name='結果')
```
